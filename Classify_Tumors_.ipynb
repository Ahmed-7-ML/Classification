# Importing Dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

# Data Preparation & Processing

```python
cancer_df = pd.read_csv("/content/breast-cancer.csv")
cancer_df.head()
```

```python
cancer_df.shape
```

```python
cancer_df.info()
```

```python
cancer_df.describe()
```

```python
cancer_df.isna().sum()
```

```python
le = LabelEncoder()
cancer_df["diagnosis"] = le.fit_transform(cancer_df["diagnosis"])
cancer_df["diagnosis"]
```

```python
cancer_df["diagnosis"].value_counts()
# 1 --> Benign
# 0 --> Malignant
```

```python
cancer_df.groupby("diagnosis").mean()
```

# Feature Extraction
## Define Features & Target

```python
X = cancer_df.drop(columns=["diagnosis", "id"], axis=1).values
y = cancer_df["diagnosis"].values
```

# Split DataFrame into Training Data & Testing Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

# Build a Model & Train it

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
logreg.fit(X_train, y_train)
```

# Model Evaluation on Training Data

```python
# Accuracy of Training Data
y_pred_train = logreg.predict(X_train)
```

```python
cm = confusion_matrix(y_train, y_pred_train)
cr = classification_report(y_train, y_pred_train)
acc = accuracy_score(y_train, y_pred_train)
```

```python
print(f"Confusion Matrix:\n{cm}")
```

```python
print(f"Classification Report:\n{cr}")
```

```python
print(f"Accuracy : {acc * 100:.2f} %")
```

# Model Evaluation on Testing Data

```python
# Accuracy of Testing Data
y_pred = logreg.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
```

```python
print(f"Confusion Matrix:\n{cm}")
```

```python
print(f"Classification Report:\n{cr}")
```

```python
print(f"Accuracy : {acc * 100:.2f} %")
```
